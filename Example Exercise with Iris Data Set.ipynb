### Let us go in point by point manner

I think I know how to convert the iris data from r to excel. But I will figure out how to convert excel to r. **Could we convert R file to excel file?**
Yes we can using the *writeXLS* R package. BTW, if you get the error of **Error in library(WriteXLS): there is no package called ‘WriteXLS’** then don't forget to install the package with install.packages("WriteXLS"). Below is the script for writing iris data set in R. The iris.xls file will be in your directory of your working R

```r
library(WriteXLS)
data(iris)
df <- iris
WriteXLS(df, ExcelFileName = "iris.xls")
```

## How to find those four attributes in R?

In the iris data frame there were 5 columns, the column containing *numerics* are attributes. You can call them, features, descriptors or attributes. And the last column is the class. So at first, we use the *names* function to get the column names and subset those with column. You can find those four attributes in multiple ways one way is to look at the head of the column or subset them. Anyway you want to look at it. 

```r
names(df)
attributes  <- df[, c("Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width")]
head(df)
```

### how to find the mean of those 4 attributes in R

You can get the mean of those 4 attributes in R by using apply function in R. Apply family function are really useful because they are vectorized which speed up things. They are much more expressive than of *for loop* in R. 

```r
mean <- apply(attributes, 2, mean)
mean
```

### Let us create some Density plots
In here, we will use *plot* function. In that function, arguments **ylab** is Y label, **xlab** is X label and **main** is title of the plot.

```r
dens.pw = density(iris$Petal.Width)
plot(dens.pw, ylab = "f(width)", xlab = "width", main = "Density of Petal Width")
```

Let us create an density plot for each class of iris flowers.


```r
head(df, 4)
```

```r
density.setosa = density(df$Petal.Length[df$Species == "setosa"])
density.versicolor = density(df$Petal.Length[df$Species == "versicolor"])
density.virginica = density(df$Petal.Length[df$Species == "virginica"])
```

```r
plot(density.setosa, ylab="f(length)", xlab="length (cm)", main="Density plot of Petal Legnths", 
     xlim = c(0,8), lwd=4, col="red")
lines(density.versicolor, col="blue", lwd=4)
lines(density.virginica, col="green", lwd=4)
legend(4, 2.5, c("Setosa", "Versicolor", "Virginica"), lwd=rep(4,3), 
       col=c("red", "blue", "green"))
```

### Quantile Plot

```r
ps = seq(0,1,length=25)
quantile.setosa = quantile(df$Petal.Length[df$Species == "setosa"], probs=ps)
quantile.versicolor = quantile(df$Petal.Length[df$Species == "versicolor"], probs=ps)
quantile.virginica = quantile(df$Petal.Length[df$Species == "virginica"], probs=ps)
```

**red color for setosa**
**blue color for versicolor**
**green color for virginica**

```r
plot(ps, quantile.setosa, xlab = "p", ylab = "length (cm)", main = "Quantiles for Petal Length", 
     type="l", lwd = 4, col="red", ylim=c(0,8))
lines(ps, quantile.versicolor, col = "blue", lwd = 4)
lines(ps, quantile.virginica, col = "green", lwd = 4)
```

## Now let us use ggplot2 to create plot
People use ggplot intead of base plot in R is because it gives you power to adjust complex figures and makes publication quality figures. 

### A basic scatter plot

```r
library(ggplot2)
ggplot(data = df, aes(x = Sepal.Length, y = Sepal.Width)) + geom_point()
```

### You can assemble plot in pieces


```r
myplot <- ggplot(data = df, aes(x = Sepal.Length, y = Sepal.Width))
myplot + geom_point()
```

You can change the color of size of the point

```r
ggplot(data = df, aes(x = Sepal.Length, y = Sepal.Width)) + geom_point(size = 3)
```

### Beautify the plot with color based on Species of the flower

```r
ggplot(data = df, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) + geom_point(size = 3)
```

### Beautify the plot with color and shape based on Species of the flower

```r
ggplot(data = df, aes(x = Sepal.Length, y = Sepal.Width, color = Species, shape = Species)) + geom_point(size = 3)
```

### Now let us create box plot using ggplot2 R package

```r
ggplot(df, aes(x = Species, y = Sepal.Length)) + geom_boxplot()
```

Make histogram for each class of Species

```r
ggplot(df, aes(Sepal.Length)) + geom_histogram() + facet_grid(Species ~.)
```

Change the histogram plot one column to get one row along with three columns

```r
ggplot(df, aes(Sepal.Length)) + geom_histogram() + facet_grid(. ~ Species)
```

You can add the regression line

```r
ggplot(df, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) + geom_point() + stat_smooth(method = "lm")
```

```r
ggplot(data = df, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) + 
geom_point(size = 3) + scale_colour_manual(values = c("red", "blue", "yellow"))
```
